The image "tire" is really blurred, we want to see more details about this tire, please use the image processing technique you have learned to show a clear tire image.

```python
# download image from url
import urllib
url = 'https://lin-chen-va.github.io/pythonforkids/Week-2/tire.tif'
image=urllib.URLopener()
image.retrieve(url,'tire.tif')

# open image
from skimage import io
tire_img = io.imread('tire.tif')

# display image
import matplotlib.pyplot as plt
plt.imshow(tire_img)
plt.show()

# add your code below
```
